# Creating Line Charts

```python
#Sampling data to create line chart
np.random.seed(4)
```

```python
#Spliting the dataframe at a 70/30 ratio and create new list holding True/False values to test np.random.rant()<=0.7
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
dev
```

```python
# Store 70% of the sample in the dataframe big
big = ords_prods_merge[dev]
```

```python
# Store 30% of the sample in the dataframe small
small = ords_prods_merge[~dev]
```

```python
# Check the length of dataframe
len(ords_prods_merge)
```

```python
#Check the length of sample subsets
len(big) + len(small)
```

```python
#Select columns necessary to create line chart
df_2= small[['orders_day_of_week','prices']]
```

```python
# Create line chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week', y = 'prices')
```

```python
# Compare line chart of subset with line chart of entire dataframe
df_4 = ords_prods_merge[['orders_day_of_week','prices']]
```

```python
# Create Line chart of entire dataframe
line_2 = sns.lineplot(data = df_4, x = 'orders_day_of_week', y = 'prices')
```
