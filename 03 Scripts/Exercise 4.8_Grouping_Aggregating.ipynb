### Create a loyalty flag using loc

```python
#Step 4 :Create a loyalty flag for existing customers using the transform() and loc() functions.
ords_prods_merge['max_order']=ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
 # Deriving columns with loc()
ords_prods_merge.loc[ords_prods_merge['max_order']>40,'loyalty_flag']='Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order']<=40)&(ords_prods_merge['max_order']>10),'loyalty_flag']='Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order']<=10,'loyalty_flag']='New customer'
```

```python
ords_prods_merge.head(15)
```

```python
#Check Frequency of 'loyalty_flag'column
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
#Step 5:Multiple Aggregation
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max','sum']})
```

From above output ,prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
#Step 6: Create a spending flag for each user based on the average price across all their orders using given criteria
ords_prods_merge['avg_price']=ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price']<10,'spending_flag']='Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price']>=10,'spending_flag']='High Spender'
```

```python
ords_prods_merge.head(15)
```

```python
#Check frequency of 'spending_flag' column
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#Step7:Create frequency flag that marks the regularity of a user’s ordering behavior to median in the “days_since_prior_order” column.
ords_prods_merge['median_freq'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_freq'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_freq'] > 10) & (ords_prods_merge['median_freq'] <= 20), 'order_freq_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_freq'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```
