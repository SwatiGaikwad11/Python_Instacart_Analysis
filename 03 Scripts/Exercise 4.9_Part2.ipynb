#  Line Charts

```python
#Spliting the dataframe at a 70/30 ratio and create new list holding True/False values to test np.random.rand()<=0.7
dev = np.random.rand(len(df_ords_prods_customers)) <= 0.7
```

```python
dev
```

```python
# Store 70% of the sample in the dataframe big
big = df_ords_prods_customers[dev]
```

```python
# Store 30% of the sample in the dataframe small
small = df_ords_prods_customers[~dev]
```

```python
# Check the length of dataframe
len(df_ords_prods_customers)
```

```python
#Check the length of sample subsets
len(big) + len(small)
```

```python
#Select columns necessary to create line chart
df_2= small[['order_hour_of_day','prices']]
```

```python
df_2.head()
```

```python
# Step 5: Create line chart
line = sns.lineplot(data = df_2, x = 'order_hour_of_day', y = 'prices')
```

```python
#Step 6: Create a line chart exploring the connections between age and number of dependents:
line_2 = sns.lineplot(data = df_4, x = 'age', y = 'number_of_dependants')
```
